```python
import re
import pandas as pd
```

```python
f = open('WhatsApp Chat with IIT Raisen Road.txt','r',encoding = 'utf-8')
```

```python
data = f.read()
```

```python
print(data)
```

```python
pattern = r'\d{1,2}/\d{1,2}/\d{2,4},\s\d{1,2}:\d{2}\s?[APM]*\s-\s'
```

```python
messages = re.split(pattern, data)[1:]
messages
```

```python
data = data.replace("\u202f", " ") 
```

```python
dates = re.findall(pattern, data)
dates
```

```python
df = pd.DataFrame({'user_message': messages, 'message_date': dates})

df['message_date'] = df['message_date'].str.rstrip(' - ')

df['message_date'] = pd.to_datetime(df['message_date'], format='%m/%d/%y, %I:%M %p', errors='coerce')

df.rename(columns={'message_date': 'date'}, inplace=True)
df.head()
```

```python
df.shape
```

```python
print(df.columns)  # Check actual column names
```

```python
users = []
messages = []
for message in df['user_message']:
    entry = re.split(r'([\w\W]+?):\s', message)
    if entry[1:]:  # user name
        users.append(entry[1])
        messages.append(" ".join(entry[2:]))
    else:
        users.append('group_notification')
        messages.append(entry[0])

df['user'] = users
df['message'] = messages
df.drop(columns=['user_message'], inplace=True)

df.head()
```

```python
df['year'] = df['date'].dt.year
```

```python
temp = df[df['user'] != 'group_notification']
temp = temp[temp['message'] != '<Media omitted>\n']
```

```python
f = open('stop_hinglish.txt','r')
stop_words = f.read()
print(stop_words)
```

```python
words = []

for message in temp['message']:
    for word in message.lower().split():
         if word not in stop_words:
            words.append(word)
```
